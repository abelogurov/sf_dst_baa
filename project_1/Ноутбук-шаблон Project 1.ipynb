Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

```python
# Функция преобразования значений признака "Образование".
def get_education(education):
    
    # Создаем список по разделетилю - пробел.
    education_list = education.split(' ')
    # Обрезаем список и заносим в переменную "education_type",
    # значения относящиеся к категории образования.
    education_type = education_list[0:2]    
    # Проверяем условием, если вхождение слова "образование",
    # если возвращается истина, то заносим первое значние,
    # в переменную "education_type".
    if education_type[1] == 'образование':
        education_type = education_list[0]        
    # Подготовка значений к очистке от символов, знаков препинания.
    education_type = str(education_type)# Перевод к типу строка.
    education_type = education_type.lower()# Привести символы к нижнему регистру.
    ed_simbols_list = ("'", '[', ']', ',')# Список символов и знаков препинания.   
    # Очистка от спецсимволов, знаков препинания.
    for ed_simbols in ed_simbols_list:
        education_type = education_type.replace(ed_simbols, '')
    
    # Возвращаем значения функции "get_education"
    return education_type

# Передаем на ввод фунукции "get_education" значения из признака "Образование и ВУЗ".
educations_types = hh_df['Образование и ВУЗ'].apply(get_education)

# Создаём новый признак "Образование", заполняем значениями функции "get_education".
hh_df['Образование'] = hh_df['Образование и ВУЗ'].apply(get_education)

# Удаляем не нужный признак "Образование и ВУЗ".
hh_df = hh_df.drop('Образование и ВУЗ', axis=1) 

# Выводим уникальные категории признака "Образование и ВУЗ".
hh_df['Образование'].unique()

### Изменение 1. 
### Копируем рабочию таблицу "hh_df" в основную таблицу "hh_data".
hh_data = hh_df.copy()
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Функция преобразования значений признака "Опыт работы (месяц)".
def get_experience(arg):
    # Проверяем условием, наличие пустых значений или слов "Не указано",
    # если значение истина, то возвращаем None
    if arg is np.nan or arg == 'Не указано':
        return None
    year_words=['год', 'года', 'лет']# Создаем список слов отвечающих за понятие год.
    month_words=['месяц', 'месяца', 'месяцев']# Создаем список слов отвечающих за понятие месяц. 
    arg_splitted = arg.split(' ')[:7]# Обрезаем список по 8 символ (не включая 8 символ).
    years = 0# Присваиваем 0 переменной "year", для занесения в переменную значений год.
    months = 0# Присваиваем 0 переменной "month", для занесения в переменную значений месяц.
    # Создаем индексированный цикл по ранее обрезаному списку.
    for index, item in enumerate (arg_splitted):
        if item in year_words:# Проверяем условием, вхождения значений в список "year_words".
            years = int(arg_splitted[index-1])# Результат истина, заносим в переменную "years" значение по индексу -1.
        if item in month_words:# Проверяем условием, вхождения значений в список "month_words".
            months = int(arg_splitted[index-1])# Результат истина, заносим в переменную "months" значение по индексу -1
    return int(years*12 + months)# Возвращаем значение сумму лет в месяцах с месяцами.

# Создаем новый признак "Опыт работы (месяц)"" на входе значений из функции "get_experience".
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_experience)

# Удаляем не нужный признак "Опыт работы".
hh_df = hh_df.drop('Опыт работы', axis=1)

### Изменение 3. 
### Копируем рабочию таблицу "hh_df" в основную таблицу "hh_data".
hh_data = hh_df.copy()
```

```python
# Ответ на задание 3.3.
# В переменную "median_experience" заносим медианное занчение признака "Опыт работы (месяц)".
median_experience = round(hh_data['Опыт работы (месяц)'].median())
print(f'Медианный опыт работы: {median_experience} месяцев.')
```

```python
# Функция преобразования значений признака "Город, переезд, командировки".
def get_city(arg):
    # Создаем список городов-миллионников.
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород',
                      'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 
                      'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград'
                     ]
    city = arg.split(' , ')[0]# Создаем список по разделителю " , " и обрезаем по первому значению.
    # Проверяем условием вхождение названия городов.
    if (city == 'Москва') or (city == 'Санкт-Петербург'):
        return city# Результат истина, возвращаем значение "Москва" или "Санкт-Петербург".
    elif city in million_cities:# Продолжаем условие по списку "million_cities".
        return 'город миллионник'# Результат истина, возвращаем занчение "город миллионник".
    else:
        return 'другие'# Иначе, звозвращаем начение "другие".
    
# Создаем новый признак "Город" на входе значений из функции "get_city".
hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(get_city)

### Изменение 4. 
### Копируем рабочию таблицу "hh_df" в основную таблицу "hh_data".
hh_data = hh_df.copy()

# Выводим уникальные значения признака "Город".
print(hh_data['Город'].unique())
```

```python
# Преобразование признака "Обновление резюме" в формат Datetime, только по дате.
def get_date_time(arg):
    date_time = arg.split(' ')[0]# Создаем список по разделителю - пробел, обрезаем список по первому значению.
    return date_time# Возвращаем полученные значения.

# Заносим в переменную "dates" значения на вход в функцию "get_date_time". 
dates = hh_df['Обновление резюме'].apply(get_date_time)

# Перезаписываем признак "Обновление резюме" значениями типа дата.
hh_df['Обновление резюме'] = dates

# Преобразование столбеца "Обновление резюме" в формат "datetime".
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], infer_datetime_format=True)

### Изменение 8. 
### Копируем рабочию таблицу "hh_df" в основную таблицу "hh_data".
hh_data = hh_df.copy()
```

```python
# Функция преобразования валют признака "ЗП".
def get_iso_currency(arg):
    # Создаем список по разделителю - пробел, обрезаем список по второму значению.
    iso_currency = arg.split(' ')[1]
    # Проверяем условием аргумент, истина возращает ISO тип валюты.
    if iso_currency == 'руб.':
        iso_currency = 'RUB'
    elif iso_currency == 'бел.руб.':
        iso_currency = 'BYN'
    elif iso_currency == 'грн.':
        iso_currency = 'UAH'
    elif iso_currency == 'сум':
        iso_currency = 'UZS'   
    return iso_currency# Возвращаем полученные значения.

# Функция преобразование признкака "ЗП".
def get_wage(arg):
    wage = arg.split(' ')[0]# Создаем список по разделителю - пробел, обрезаем список по первому значению.
    return int(wage)# Возвращаем полученные значения.

hh_df['ISO валюта'] = hh_df['ЗП'].apply(get_iso_currency)# Создаём новый признак "ISO валюта".
hh_df['ЗП'] = hh_df['ЗП'].apply(get_wage)# Создаем преобразованный признак "ЗП".

hh_df = pd.merge(hh_df, er_df, how='left')# Объединение таблиц "hh_df" и "er_df", по совпадающим признакам таблицы "hh_df".
hh_df['close'] = hh_df['close'].fillna(1)# Замена значений "nan", в признакае "close", на 1 (для умножения на курс).
hh_df['proportion'] = hh_df['proportion'].fillna(1)# Замена значений "nan", в признаке, "proportion" на 1 (для деления на пропорцию).

hh_df['ЗП (руб.)'] = (hh_df['ЗП'] * hh_df['close'])/hh_df['proportion'] #Переводим значения признака "ЗП" в рубли.

hh_df = hh_df.drop(['ISO валюта', 'close', 'proportion'], axis=1) #Удаляем не нужные признаки: 'ISO валюта', 'close', 'proportion'.
hh_df = hh_df.drop('ЗП', axis=1)# Удаляем не нужный признак "ЗП".

### Изменение 9. 
### Копируем рабочию таблицу "hh_df" в основную таблицу "hh_data".
hh_data = hh_df.copy()
```

```python
# Ответ на задание 5.6.

def outliers_z_age(data, feature, left=3, right=3, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned

outliers, data = outliers_z_age(hh_cl, 'Возраст', left=3,  right=4, log_scale=True)
print(f'Методом z-отклонений было обнаружено: {outliers.shape[0]} выброса.')
print(outliers['Возраст'].unique())
```
